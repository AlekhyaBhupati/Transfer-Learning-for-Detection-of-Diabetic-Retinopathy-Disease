```python
#dictionary for level of images
level = {}
for i in range(df.shape[0]):
  level[df['image'][i]] = df['level'][i]
```

```python
images = []
label = []
```

```python
# reading images
count = [0,0,0,0,0]
for filename in os.listdir("train/"):
  img = cv2.imread(os.path.join("train/",filename))
  img = cv2.resize(img,(32,32))
  if img is not None:
    if count[level[filename[:-5]]]<2500:
      images.append(img)
      label.append(level[filename[:-5]])
      count[level[filename[:-5]]]+=1
```

```python
pd.Series(count).plot.bar()
```

```python
X,y = images, label
```

```python
X = np.array(X)
y = np.array(y)
```

```python
#splitting data into training and testing data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

# **DenseNet201**

```python
#model
import tensorflow as tf
from tensorflow.keras.layers import Dense, Dropout, Flatten, Conv2D, MaxPool2D,GlobalAveragePooling2D
from tensorflow.keras.optimizers import Adam
from tensorflow.keras import applications
from tensorflow.keras.models import Sequential,Model,load_model
img_height,img_width = 32,32
num_classes = 5
base_model = applications.densenet.DenseNet201(weights= 'imagenet', include_top=False, input_shape= (img_height,img_width,3))
base_model.trainable = False
x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dropout(0.7)(x)
predictions = Dense(num_classes, activation= 'softmax')(x)
model = Model(inputs = base_model.input, outputs = predictions)
```

```python
#Optimizer
adam = tf.keras.optimizers.Adam(lr=0.0001)
model.compile(optimizer= adam, loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
# model layers
model.summary()
```

```python
#flowchart of model
from tensorflow.keras.utils import plot_model
plot_model(model)
```

```python
#training
history = model.fit(X_train, y_train, validation_data=(X_test,y_test), epochs = 40, batch_size = 32)
```

```python
#testing
loss, acc = model.evaluate(X_test,y_test)
print("loss: ",loss)
print("accuracy: ",acc)
```

```python
#Classification Matrices
from sklearn.metrics import accuracy_score, classification_report
y_pred  =np.argmax(model.predict(X_test),axis=1)
print('Accuracy on Test Data: %2.2f%%' % (accuracy_score(y_test, y_pred)))
print(classification_report(y_test, y_pred))
```

```python
#confusion Matrix
import seaborn as sns
from sklearn.metrics import confusion_matrix
sns.heatmap(confusion_matrix(y_test, y_pred),annot=True, fmt="d", cbar = False, cmap = plt.cm.Blues, vmax = X_test.shape[0]//16)
```
